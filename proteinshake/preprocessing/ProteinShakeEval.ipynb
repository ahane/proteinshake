```python
import json
import pandas as pd
from matplotlib import pyplot as plt
from pandas import DataFrame, Series
%pylab inline
```

```python
def df_from_radio(json):
    distances = [tuple(x) for x in radiojson['connections']]
    distdf = DataFrame(distances, columns=['edge', 'distance'])
    dist_with_index = distdf.set_index('edge')
    return dist_with_index
```

```python
def df_from_ProteinShake(json, name):
    predictions = [x.items()[0] for x in json['predictions']]
    preddf = DataFrame(predictions, columns=['edge', name])
    pred_with_index = preddf.set_index('edge')
    return pred_with_index
```

```python
def df_with_tp_rates(df, name, threshold=8):
    df_copy = df.copy()
    all_predicted = df_copy[df_copy[name] > -1]
    all_predicted_and_measured = all_predicted[all_predicted['distance'] > -1]
    all_predicted['dist_under_thres'] = all_predicted['distance'].apply(lambda x: x <= threshold)
    sorted_by = all_predicted.sort(name, ascending=False)
    
    sorted_by['positives'] = sorted_by['dist_under_thres'].apply(lambda x: 1 if x else 0)
    sorted_by['cuml_positives'] = sorted_by['positives'].cumsum(axis=0)
    sorted_by['positives']
    
    sorted_by['count_helper'] = sorted_by['edge'].apply(lambda x: 1)
    sorted_by['predictions_count'] = sorted_by['count_helper'].cumsum(axis=0)
    del sorted_by['count_helper']
    
    sorted_by['tp_rate'] = sorted_by['cuml_positives'] / sorted_by['predictions_count']
    
    return sorted_by
```

```python
with open('pf14/MI.json') as f:
    mijson = json.loads(f.read())
with open('pf14/PF00014all.json') as f:
    radiojson = json.loads(f.read())
with open('pf14/7_699047304992377E-100_9_3_1.json') as f:
    cdjson = json.loads(f.read())
with open('pf14/1_992726023628825E-95_820988186885298_3_1.json') as f:
    cd2json = json.loads(f.read())
```

```python
dfs = [df_from_radio(radiojson), df_from_ProteinShake(mijson, 'mi'), df_from_ProteinShake(cdjson, 'cd'), df_from_ProteinShake(cd2json, 'cd2')]
alldf = pd.concat(dfs, axis=1)
alldf['edge'] = alldf.index
alldf = alldf.fillna(-1)
```

```python
alldf.sort('mi', ascending=False)
```

```python
mi_df = df_with_tp_rates(alldf, 'mi')
cd_df = df_with_tp_rates(alldf, 'cd')
cd2_df = df_with_tp_rates(alldf, 'cd2')
```

```python
def plot_multiple(dfs):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    num_contacts = dfs[0][0]['positives'].sum()
    for df, name in dfs:
        plt.plot(df['predictions_count'], df['tp_rate'], label=name)
        plt.xscale('log')
        
        final_value = round(df['tp_rate'][num_contacts], 4)
        ax.annotate(final_value, xy=(num_contacts, final_value))
    
    plt.xlabel("Number of Contacts")
    plt.xlim(1, num_contacts)
    plt.ylabel("TP-Rate until n-th Prediction")
    plt.legend()
```

```python
plot_multiple([(cd_df, 'Contrastive Divergence'), (mi_df, 'Mututal Information'), (cd2_df, 'Contrastive 2')])
```
